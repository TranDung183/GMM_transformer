```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.mixture import GaussianMixture
import torch
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt

# Thư viện Informer (ví dụ Informer2020, cần cài riêng)
# from informer import Informer  # Giả sử bạn đã cài thư viện Informer

# 1. Load dữ liệu và chuẩn hóa
df = pd.read_excel('yahoo_data.xlsx')
df.columns = df.columns.str.strip().str.replace('*', '', regex=False).str.replace(' ', '_').str.lower()
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

scaler_close = MinMaxScaler()
df['close_scaled'] = scaler_close.fit_transform(df[['close']])
scaler_volume = MinMaxScaler()
df['volume_scaled'] = scaler_volume.fit_transform(df[['volume']])

# 2. GMM phân cụm
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(df[['close_scaled', 'volume_scaled']])
df['gmm_cluster'] = gmm_labels.astype(str)  # Chuyển thành string để làm categorical

# 3. Chuẩn bị dữ liệu cho Informer
seq_len = 60  # Input sequence length
pred_len = 30  # Dự đoán 30 bước

class TimeSeriesDataset(Dataset):
    def __init__(self, data, seq_len, pred_len):
        self.data = data
        self.seq_len = seq_len
        self.pred_len = pred_len
        self.features = data[['close_scaled', 'volume_scaled']].values
        self.clusters = data['gmm_cluster'].astype(int).values  # Chuyển lại thành int cho model

    def __len__(self):
        return len(self.data) - self.seq_len - self.pred_len + 1

    def __getitem__(self, idx):
        x_seq = self.features[idx:idx+self.seq_len]
        x_cluster = self.clusters[idx:idx+self.seq_len]
        y_seq = self.features[idx+self.seq_len:idx+self.seq_len+self.pred_len, 0]  # Giá close_scaled
        return torch.tensor(x_seq, dtype=torch.float32), torch.tensor(x_cluster, dtype=torch.long), torch.tensor(y_seq, dtype=torch.float32)

dataset = TimeSeriesDataset(df, seq_len, pred_len)

train_size = int(len(dataset)*0.8)
test_size = len(dataset) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)

# 4. Xây dựng mô hình Informer (giả định bạn có mô hình Informer sẵn)
# Dưới đây là một mẫu mô hình giả định (bạn cần thay bằng mô hình Informer thật)
import torch.nn as nn

class SimpleInformer(nn.Module):
    def __init__(self, input_dim, cluster_dim, hidden_dim, output_dim, seq_len, pred_len):
        super(SimpleInformer, self).__init__()
        self.embedding = nn.Linear(input_dim, hidden_dim)
        self.cluster_embedding = nn.Embedding(10, hidden_dim)  # embedding cho cluster (giả định tối đa 10 cluster)
        self.transformer = nn.Transformer(d_model=hidden_dim, num_encoder_layers=2, num_decoder_layers=2)
        self.fc_out = nn.Linear(hidden_dim, output_dim)
        self.seq_len = seq_len
        self.pred_len = pred_len

    def forward(self, x, clusters):
        # x shape: (batch, seq_len, input_dim)
        # clusters shape: (batch, seq_len)
        x_embed = self.embedding(x)  # (batch, seq_len, hidden_dim)
        c_embed = self.cluster_embedding(clusters)  # (batch, seq_len, hidden_dim)
        x_combined = x_embed + c_embed  # cộng embedding
        x_combined = x_combined.permute(1, 0, 2)  # Transformer yêu cầu (seq_len, batch, feature)

        # Đầu vào cho Transformer (giả định decoder dùng zero tensor)
        tgt = torch.zeros(self.pred_len, x.size(0), x_combined.size(2)).to(x.device)

        out = self.transformer(x_combined, tgt)  # (pred_len, batch, hidden_dim)
        out = self.fc_out(out)  # (pred_len, batch, output_dim)
        out = out.permute(1, 0, 2)  # (batch, pred_len, output_dim)
        return out.squeeze(-1)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = SimpleInformer(input_dim=2, cluster_dim=1, hidden_dim=64, output_dim=1, seq_len=seq_len, pred_len=pred_len).to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# 5. Huấn luyện mô hình
epochs = 30
model.train()
for epoch in range(epochs):
    total_loss = 0
    for x_seq, x_cluster, y_seq in train_loader:
        x_seq, x_cluster, y_seq = x_seq.to(device), x_cluster.to(device), y_seq.to(device)
        optimizer.zero_grad()
        outputs = model(x_seq, x_cluster)
        loss = criterion(outputs, y_seq)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(train_loader):.5f}")

# 6. Dự đoán và đánh giá
model.eval()
preds = []
trues = []
with torch.no_grad():
    for x_seq, x_cluster, y_seq in test_loader:
        x_seq, x_cluster, y_seq = x_seq.to(device), x_cluster.to(device), y_seq.to(device)
        outputs = model(x_seq, x_cluster)
        preds.append(outputs.cpu().numpy())
        trues.append(y_seq.cpu().numpy())

preds = np.concatenate(preds)
trues = np.concatenate(trues)

# Chuyển ngược về giá trị gốc
preds_inv = scaler_close.inverse_transform(preds)
trues_inv = scaler_close.inverse_transform(trues)

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
plt.plot(trues_inv.flatten(), label='Thực tế')
plt.plot(preds_inv.flatten(), label='Dự đoán')
plt.title('Dự báo giá đóng cửa với GMM + Informer')
plt.legend()
plt.show()
```
